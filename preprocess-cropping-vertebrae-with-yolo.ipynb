# Save Images for YOLOv5 training
I found it easier to save the CT slices as jpegs for training Ultralytics YOLO.

```python
# patients with segmentation data
niis = os.listdir('/kaggle/input/rsna-2022-cervical-spine-fracture-detection/segmentations')
train_df = pd.read_csv('/kaggle/input/3-channel-preprocessed-dataset/3_channel_train_df1.csv')
#pts = [re.search('(1.2.826.0.1.3680043.)([0-9]*)(?=.nii)', filename).group(0) for filename in niis]
pts = np.unique(list(train_df['StudyInstanceUID']))
#pts = np.unique([each_img[:-4].split('_')[0] for each_img in os.listdir('/kaggle/input/3-channel-preprocessed-dataset')])

# Save slices corresponding to each yolo_coord txt file
train_images = '/kaggle/input/3-channel-preprocessed-dataset/prep_train'

yolo_slices = os.path.join(os.getcwd(), 'yolo_slices')
if not os.path.exists(yolo_slices):
    os.mkdir(yolo_slices)
```

```python
for pt in tqdm(pts):
    slice_nums_ = len(glob(f'/kaggle/input/rsna-2022-cervical-spine-fracture-detection/train_images/{pt}/*'))
    slice_nums = [int(re.search('([0-9]*)', str(filename)).group(0)) for filename in range(slice_nums_)]
    slice_nums = sorted(slice_nums)
    pt_CT = os.path.join(train_images, str(pt)) 
    
    for slice_ in slice_nums:
        uid_id = f'{pt}_{slice_}'
        img_path = f'{pt_CT}_{slice_}'
        img = np.load(f'{img_path}.npz')['arr_0']

        imgs_savepath = os.path.join(yolo_slices, f"{str(pt)}_{slice_}.jpg")
        cv2.imwrite(imgs_savepath, img)
```
